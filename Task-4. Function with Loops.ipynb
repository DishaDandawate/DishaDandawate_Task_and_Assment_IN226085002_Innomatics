### Problem 1: Mobile Recharge Validation System

```python
# function to check recharge
def recharge_check(amount):
    valid_plans=[199, 299, 399, 599]
    if amount < 50:
        print("Recharge must be at least 50")
        return False
    if amount not in valid_plans:
        print("Invalid plan. Choose from 199, 299, 399, 599")
        return False
    print("Recharge Successful")
    return True
# retry using while loop
while True:
    amnt = int(input("Enter recharge amount: "))
    if recharge_check(amnt):
        break   # stop loop if correct
```

### Problem 2: Inventory Reorder Alert System

```python
# dictionary = product : quantity
products = {
    "Pen": 10,
    "Book": 20,
    "Pencil": 5,
    "Eraser": 18
}
# function
def check_stock(items):
    for name, quantity in items.items():   # go one by one
        if quantity < 15:
            print(name, ": Reorder Alert")
        else:
            print(name, ": Stock OK")
check_stock(products)
```

###  Problem 3: Student Result Processing System

```python
def result(marks):
    total = 0
    # add all marks
    for m in marks:
        total += m
    avg = total / len(marks)
    print("Average =", avg)
    if avg >= 50:
        print("Pass")
    else:
        print("Fail")
# input
student_marks = [62, 48, 70, 51]
result(student_marks)
```

### Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(km, peak):
    fare = 50 + (12 * km)   # basic fare
    if peak == "yes":
        fare = fare + (fare * 0.25)  # add 25%
    print("Total Fare =", fare)
# retry loop
while True:
    km = int(input("Enter distance (km): "))
    peak = input("Peak hour? [yes/no]: ")
    calculate_fare(km, peak)
    again = input("Calculate again? [yes/no]: ")
    if again == "no":
        break
```

### Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_check(days):
    present = 0
    # count P
    for d in days:
        if d == "P":
            present += 1
    percent = (present/len(days)) * 100
    print("Attendance =", percent, "%")
    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
data = ["A","P","P","P","A","A","P","P"]
print(attendance_check(data))
```

### Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        print("Password too short")
        return
    # check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")
password = input("Enter password: ")
check_password(password)
```
