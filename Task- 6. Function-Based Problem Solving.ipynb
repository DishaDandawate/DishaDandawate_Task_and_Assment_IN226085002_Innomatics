### Problem Statement 1:Smart Parking Lot Management System

```python
# Function to manage parking lot
def parking_system(capacity, logs):
    parked = 0
    peak = 0
    # Loop through logs
    for log in logs:
        if log == "IN":
            parked = parked + 1
        elif log == "OUT":
            if parked > 0:
                parked = parked - 1
        # Check peak usage
        if parked > peak:
            peak = parked
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    # Check capacity
    if parked > capacity:
        print("Parking Status: Full / Exceeded")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_system(capacity, logs)
```

### Problem Statement 2:Online Food Delivery Time Estimator

```python
# Function to calculate delivery time
def delivery_time(distance, traffic, weather):
    # Base time = 5 minutes per km
    time = distance * 5
    # Traffic condition
    if traffic == "High":
        time = time + 15
    elif traffic == "Medium":
        time = time + 10
    else:
        time = time + 5
    # Weather condition
    if weather == "Rainy":
        time = time + 10
    elif weather == "Stormy":
        time = time + 20
    print("Estimated Delivery Time:", time, "minutes")
delivery_time(8, "High", "Rainy")
```

### Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
# Function to analyze seats
def seat_analysis(total_seats, booked_list):
    booked = 0    
    # Count booked seats
    for seat in booked_list:
        if seat == 1:
            booked = booked + 1
    percentage = (booked / total_seats) * 100
    print("Occupancy:", percentage, "%")
    # Check status
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
total = 200
booked = [1] * 150

seat_analysis(total, booked)
```

###  Problem Statement 4:Cloud Server Load Classification System

```python
# Function to check server load
def server_load(cpu_list):
    total = 0
    # Calculate total
    for value in cpu_list:
        total = total + value
    average = total / len(cpu_list)
    print("Average CPU Load:", average, "%")
    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average >= 50 and average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu = [45, 60, 70, 85, 90]

server_load(cpu)
```

### Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
# Function to check resource usage
def resource_monitor(resources):  
    overused = []
    # Assume overuse if usage > 8 hours
    for item in resources:
        if resources[item] > 8:
            overused.append(item)
    if len(overused) > 0:
        print("Overused Resources:", overused)
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

###  Problem Statement 6: Online Event Registration Capacity Controller

```python
# Function to manage registrations
def event_registration(capacity, registrations):    
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100, 105)
```
