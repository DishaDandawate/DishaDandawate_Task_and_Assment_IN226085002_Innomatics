### Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post(likes_list):
    total = 0
    # calculate total likes using loop
    for like in likes_list:
        total = total + like
    # check status
    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total)
    
    print("Post Status:", status)
# input
likes = list(map(int, input("Enter likes separated by space: ").split()))
analyze_post(likes)
```

### Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfalls, required):
    total = 0
    # calculate total using loop
    for r in rainfalls:
        total = total + r
    avg = total / len(rainfalls)
    # condition
    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(avg))
    print("Rainfall Status:", status)
# input
rainfalls = list(map(int, input("Enter rainfall values: ").split()))
required = int(input("Enter required rainfall: "))
check_rainfall(rainfalls, required)
```

### Problem 4: Social Media – Duplicate Account Detection

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    print("Duplicate Accounts Found:", result)
# input
usernames = input("Enter usernames separated by space: ").split()
check_duplicates(usernames)
```

### Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crops(prices):
    premium = []
    regular = []
  
    for p in prices:
        if p > 2000:
            premium.append(p)
        else:
            regular.append(p)
    print("Premium Crops:", premium)
    print("Regular Crops:", regular)

# input
prices = list(map(int, input("Enter crop prices: ").split()))
premium_crops(prices)
```

### Problem 7: System Monitoring – Application Health Checker

```python
def check_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues" 
    print("Error Count:", errors)
    print("System Status:", status)
# input
errors = int(input("Enter error count: "))
check_health(errors)
```

### Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_eligibility(attendance):

    total_classes = len(attendance)
    present = 0
    for day in attendance: 
        if day == "P":
            present += 1
    percentage = (present / total_classes) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
# attendance record
records = ["P","P","A","P","P","P","A","P","P","P"]
percent, result = check_eligibility(records)
print("Attendance Percentage:", round(percent,1))
print("Exam Eligibility:", result)
```

### Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    # bill calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    # usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill < 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
# input
units = int(input("Enter units consumed: "))
bill, usage = electricity_bill(units)

print("Total Bill:", bill)
print("Usage Level:", usage)
```
